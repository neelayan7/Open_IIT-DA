## Feature Engineering

```python
New_parameter = np.array(x['Customer'])
for i in range(len(New_parameter)):
    New_parameter[i] = New_parameter[i][0:2]
x['New Parameter'] = New_parameter

temp = x.groupby('New Parameter')['Customer Lifetime Value'].mean()
```

```python
clv_by_id = np.array(range(x.shape[0]))
clv_by_id = clv_by_id.astype(float)
for i in range(x.shape[0]) :
    clv_by_id[i] = temp[x['New Parameter'][i]]
x['Average ID'] = clv_by_id
```

```python
x['Present Value'] = x['Monthly Premium Auto'] * x['Months Since Policy Inception'] - x['Total Claim Amount']
```

```python
x = df.drop(['Customer Lifetime Value','Gender','Customer','State'], axis = 1 )
```

## Data Pre-Processing

```python
xx = pd.get_dummies(x)
```

```python
x.head()
```

```python
xx
```

## Splitting dataset into 2 parts-Development Sample and Validation Sample

```python
X_train, X_test, y_train, y_test = train_test_split(xx, y, test_size=0.2, random_state=0)

```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state=0,min_samples_split=20)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

## Final MAPE

```python
from sklearn.utils.validation import check_array
def mean_absolute_percentage_error(y_test, y_pred): 
    #y_test, y_pred = check_arrays(y_test, y_pred)
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(mean_absolute_percentage_error(y_test, y_pred))
```

## Final RMSE

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
